# BAR CHART

```python
import matplotlib.pyplot as plt

x = ["python","c","c++","java",]
y = [85,70,60,82]
z = [20,30,40,50]

plt.xlabel("language")
plt.ylabel("No.")
plt.title("Comparring",fontsize = 25)
c = ["y","b","m","g"]

plt.bar(x,y,width=0.1,color = c,edgecolor = "black",linewidth = 2,linestyle = "dashdot",alpha = 0.4,label = "popularty")
plt.bar(x,z,width=0.1,color = c,edgecolor = "black",linestyle = "dashdot",linewidth = 2,label = "popularty1")

plt.legend()
plt.show()
```

## parameters and uses

width - width of bar
edgecolor - for color the edge
align - for align (center,edge)
linewidth - for width of border of bar
linestyle - for border style
alpha - for brightness(0 to 1.0)
label + legend() - add label in bar ,
  or
legend("...") - add label in bar
legend(loc = "upper right") - output in upper right side 
rotation - add tilted form of text
bar(...) - vertical bar graph
barh(...) - horizental bar graph

```python
import matplotlib.pyplot as plt
import numpy as np

x = ["python","c","c++","java",]
y = [85,70,60,82]
z = [20,30,40,50]

width=0.1
p = np.arange(len(x))
p1 = [i+width for i in p]

plt.xlabel("language")
plt.ylabel("No.")
plt.title("Comparring",fontsize = 25)
c = ["y","b","m","g"]

plt.bar(p,y,width,color = c,edgecolor = "black",linewidth = 2,linestyle = "dashdot",alpha = 0.4,label = "popularty")
plt.bar(p1,z,width,color = c,edgecolor = "black",linestyle = "dashdot",linewidth = 2,label = "popularty1")

plt.xticks(p+width/2,x,rotation = 45)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = ["python","c","c++","java",]
y = [85,70,60,82]
z = [20,30,40,50]

width=0.1
p = np.arange(len(x))
p1 = [i+width for i in p]

plt.ylabel("language")
plt.xlabel("No.")
plt.title("Comparring",fontsize = 25)
c = ["y","b","m","g"]

plt.barh(p,y,width,color = c,edgecolor = "black",linewidth = 2,linestyle = "dashdot",alpha = 0.4,label = "popularty")
plt.barh(p1,z,width,color = c,edgecolor = "black",linestyle = "dashdot",linewidth = 2,label = "popularty1")

plt.yticks(p+width/2,x,rotation = 45)
plt.legend()
plt.show()
```

# Scatter Plot

```python
import matplotlib.pyplot as plt

day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
no2 = [3,2,4,5,1,6,2]

color = ["r","y","g","b","r","g","r"]
size = [150,200,104,125,120,110,117]
color = [10,49,30,29,56,20,30]

plt.scatter(day,no,c = color,s = size,alpha = 0.5,marker = "*",edgecolor = "b",linewidth = 0.5, cmap = "BrBG")
plt.scatter(day,no2,c = color,s = size,marker = "*",edgecolor = "w",linewidth = 0.5, cmap = "viridis")

plt.title("Scatter plot",fontsize = 25)
plt.xlabel("Day",fontsize = 15)
plt.ylabel("No.",fontsize = 15)
t = plt.colorbar()
t.set_label("Color Bar")

plt.show()
```

## parameters and uses

marker - use for mark in graph (.,*+)
edgecolor - for colour of edge
linewidth - for width of line
cmap - use for colour mapping
colorbar() - use for colour bar for it's colour range
t.set_label("Color Bar") - add label in color bar , here 't' is variable that contain plt.colourbar()

#  Chart

```python
import matplotlib.pyplot as plt

x = [10, 20 , 30 , 40]
y = ["c" , "c++" , "java" , "python"]
ex = [0.05,0.05,0.1,0.05]
c = ["r","b","g","y"]

plt.pie(x,labels = y,explode = ex,colors = c,autopct = "%0.1f%%",shadow = "True",radius = 1.1,labeldistance = 0.8,
        startangle = 90,textprops = {"fontsize":15},counterclock = False,wedgeprops = {"linewidth":4},center = (5,5),
        rotatelabels = True)
plt.title("Pie Chart",fontsize = 20)
plt.legend(loc = 2)

plt.show()
```

explode - used to specify the fraction of the radius with which to offset each wedge
autopct - used to display the percentage value on each slice of the pie chart.
"%0.1f%%" - formats the percentage value of each slice to one decimal place and appends a percent sign. (0.1% is after 1 decimal point)
shadow - used for shadow as 3 D format (True,False)
radius - used for specifies the radius(size) of the pie chart
labeldistance - used for distance of label(1 > dis = inside, 1 < dis = outside)
startangle - used for effective way to control the orientation of a pie chart.
textprops - * always use in dictonary - used for customize the appearance of the text labels 
counterclock - used for the direction in which the pie chart is drawn (True - normal)
wedgeprops - * always use in dictonary - use for customize the apperance of the edge of the pie
center - * alway use in tuple - use for customize the apperance of the center of the pie
rotatelabels - use for rotating label (True or False)
title() - use for label
